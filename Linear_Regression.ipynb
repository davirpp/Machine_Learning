**Atividade 2:**

Implemente o algoritmo de Regressão Linear que retorne o vetor de pesos $w=(X^TX)^{-1}X^Ty$

```python
class LinearRegression_:
    def fit(self, _X, _y):
        X = np.array(_X)
        y = np.array(_y)
        self.w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
        return self.w
     
    def predict(self, _x):
        x = np.array(_x)
        return x.dot(self.w)
     
    def getW(self):
        return self.w
```
